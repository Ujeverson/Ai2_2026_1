```python
dados = pd.Series([1, 3, 5, 7, 9], index=['a', 'b', 'c', 'd', 'e'])
print(dados)

print("O primeiro elemento é",dados[0])
print(f"O terceiro elemento é {dados[2]}.")
```

```python
dados = pd.Series([1,3,5,7,9,11,13,15], index=['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h'])
print("buscando os três primeiros elementos \n")
print(dados[:3])
print('buscando os elementos de 3 a 6 ')
print(dados[3:6])
```
